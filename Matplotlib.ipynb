1) We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:

1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

Expected Output:




```python
import numpy as np
max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
import matplotlib.pyplot as plt
months=np.arange(12)
plt.plot(months,max,"go")
plt.plot(months,min,"bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
```

```python
##Fitting it to a periodic functoin
from scipy import optimize
def yearly_temps(times,avg,ampl,time_offset):
    return(avg+ampl*np.cos((times+time_offset)*1.8*np.pi/times.max()))
res_max,cov_max=optimize.curve_fit(yearly_temps,months,max,[40,20,0])
res_min,cov_min=optimize.curve_fit(yearly_temps,months,min,[-40,20,0])
```

```python
##plotting the fit
days=np.linspace(0,12,num=365)

plt.figure()
plt.plot(months,max,"go")
plt.plot(days,yearly_temps(days,*res_max),"m-")
plt.plot(months,min,"bo")
plt.plot(days,yearly_temps(days,*res_min),"y-")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()
```

2) This assignment is for visualization using matplotlib:
data to use:
url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv
titanic = pd.read_csv(url)
Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic=pd.read_csv(url,skip_blank_lines=True)
```

```python
count=titanic['sex'].value_counts()
colors=['blue','orange']
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.pie(count.values.tolist(), labels=count.index.values.tolist(), colors=colors,autopct='%.1f%%')
ax1.set_title('Male/Female proportion')
```

```python
plt.figure()
category1=titanic[titanic.sex=='male'].plot.scatter('age','fare',color='blue',label='male')
titanic[titanic.sex=='female'].plot.scatter('age','fare',color='green',label='female',ax=category1)
```
