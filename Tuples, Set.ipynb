1.1 Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```python
def my_func(a,b):
    return a+b
my_list=[1,2,3,4]
def myreduce(my_func,my_list):
    x=my_list[0]
    for i in range(1,len(my_list)):
        print("New element in the list is",x)
        x=my_func(x,my_list[i])
        print("Next element in the list is",my_list[i])
        print("Sum of above elements in sequence is",x)
    return(x)
print("Result of my reduce function is",myreduce(my_func,my_list))
```

1.2 Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()

```python
my_list=[3,12,17,18,10,9,20,24,32,0]
fill_list=[]
def my_func(a):
    if a<18:
        fill_list.append(a)
        print(fill_list)
        return "The above printed list is less than 18"
def my_filter(my_func,my_list):
    for i in  range(0,len(my_list)):
        x=my_func(my_list[i])
    return(x)
result=my_filter(my_func,my_list)
print(result)
```
